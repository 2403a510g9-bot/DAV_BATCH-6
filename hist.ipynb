```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic blood sugar data for 200 patients
# (Two groups: normal and high blood sugar)
np.random.seed(10)
normal = np.random.normal(95, 10, 120)   # Normal group
high = np.random.normal(155, 15, 80)     # High sugar group
blood_sugar = np.concatenate([normal, high])

# Create histogram and KDE plot
plt.figure(figsize=(8, 5))
sns.histplot(blood_sugar, bins=20, kde=True, color='skyblue', edgecolor='black')

# Add labels and title
plt.title("Blood Sugar Levels of 200 Patients (Histogram + KDE)", fontsize=14)
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Density")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Example data (replace with your actual blood sugar data)
blood_sugar_levels = np.random.normal(120, 15, 200) # Generating 200 random data points with mean 120 and std dev 15

# Create the KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(blood_sugar_levels, fill=True)
plt.title('Density Plot of Blood Sugar Levels')
plt.xlabel('Blood Sugar Level')
plt.ylabel('Density')
plt.show()
```
